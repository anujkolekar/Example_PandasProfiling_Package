```python
fig,ax = plt.subplots()

ax.hist(df[df["Kyphosis"]=='present']["Age"], bins=15, alpha=0.5, color="blue", label="present")
#ax.hist(df[df["Kyphosis"]=='absent']["Age"], bins=15, alpha=0.5, color="green", label="absent")

ax.set_xlabel("Age")
ax.set_ylabel("Kyphosis")

fig.suptitle("Age vs. Kyphosis")

ax.legend();
```

```python
fig, ax = plt.subplots()

sns.kdeplot(df[df["Kyphosis"]=='present']["Age"], shade=True, color="blue", label="present", ax=ax)
sns.kdeplot(df[df["Kyphosis"]=='absent']["Age"], shade=True, color="green", label="absent", ax=ax)

ax.set_xlabel("Age")
ax.set_ylabel("Kyphosis")

fig.suptitle("Age vs. Kyphosis");

```

```python
fig, ax = plt.subplots()

sns.boxplot(x="Age", y="Kyphosis", data=df, orient="h", palette={'present':"blue", 'absent':"green"}, ax=ax)

ax.get_yaxis().set_visible(False)

fig.suptitle("Age vs. Kyphosis")

color_patches = [
    Patch(facecolor="blue", label='present'),
    Patch(facecolor="green", label='absent')
]
ax.legend(handles=color_patches);
```
